## Exercise 01: fundamental theorem of calculus
In this exercise, we will demonstrate the fundamental theorem of calculus by comparing the theoretical derivative with a numerical approximation using the fundamental theorem

### Newton's laws of motion

The position x of an object with an initial starting point x0 (in m) and velocity v (in m/s) after a duration t (in seconds) is given by the equation

    Equation 1 x = x0 + vt

when the velocity is changing linearly over time, this is acceleration.  The velocity after duration t (in seconds) at a linear acceleration rate a (m/s^2) and initial velocity v0 is given by the equation

    Equation 2  v = v0 + at

Therefore, combining Equations 1 and 2, the position of an object with constant linear acceleration with initial position x0=0 and initial velocity v0=0 is given by the equation

    use calculus tables 

    Equation 3 x = 1/2*at^2

## velocity as the instantaneous rate of change in position

the fundamental theorem of calculus describes the relationship between changes in the position of a function and the instantaneous slope and is given by 

    Equation 4  dx/dt = (f(t+ delta t) - f(t))/delta t
    
or restated in the variables from our motion example, where v* is the instantaneous velocity

    Equation 5  v* = (x(t+ delta t) - x(t))/delta t
    
so we have two equations to predict the instantaneous velocity, Equation 2 and Equation 5

## plot theoretical velocity

```python
t = range(0,100)
a = 0.5
v = [a*i for i in t] #Equation 2
v[:5]
```

```python
x = [a*i**2 for i in t] #Equation 3
x[:5]
```

```python
plt.plot(t, v)
```

## plot position and velocity

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(t, v, color="red")
# set x-axis label
ax.set_xlabel("duration (sec)",fontsize=14)
# set y-axis label
ax.set_ylabel("velocity (m/s)",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(t, x,color="blue")
ax2.set_ylabel("position (m)",color="blue",fontsize=14)
plt.show()
```

## approximation of velocity from F.T.C. with forward full step

```python
def position(a, t):
    x = 1/2*a*t**2  #Equation 3
    return x

def v_ftc_forward_full(a, t, deltat):
    v = 0
    if deltat != 0:
        #Fundamental theorem of calculus 
        v = 1/deltat*(position(a, t+deltat) - position(a, t)) 
    return v

def v_ftc_centered_half(a, t, deltat):
    v = 0
    if deltat != 0:
        #Fundamental theorem of calculus 
        v = 1/deltat*(position(a, t+1/2*deltat) - position(a, t-1/2*deltat)) 
    return v
```

```python
deltat = 10
v_appx = [v_ftc_forward_full(a, i, deltat) for i in t]
v_appx[:5]
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(t, v, color="red")
# set x-axis label
ax.set_xlabel("duration (sec)",fontsize=14)
# set y-axis label
ax.set_ylabel("velocity (m/s)",color="red",fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(t, x,color="blue")
ax2.set_ylabel("position (m)",color="blue",fontsize=14)

# make a plot with different y-axis using second axis object
ax.plot(t, v_appx, "m--")
plt.show()
```
